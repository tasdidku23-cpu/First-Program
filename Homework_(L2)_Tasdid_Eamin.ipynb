```python
# Task-8 (Creating two sets of color)
boys_favorite = {"blue", "black", "white", "gray", "navy blue"}
girls_favorite = {"pink", "purple", "red", "blue", "white"}

# union
total_favorite_colors = boys_favorite | girls_favorite
print(total_favorite_colors)
# intersection
common_colors = boys_favorite & girls_favorite
print(common_colors)
# difference
boys_only = boys_favorite - girls_favorite
print(boys_only)
girls_only = girls_favorite - boys_favorite
print(girls_only)
```

```python
# Task-9 (Creating function for first and last character)
def print_first_last(text):
  first_character = text[0]
  last_character = text[-1]
  print("First Character:", first_character)
  print("Last Character:", last_character)

# example 1
print_first_last("Tasdid")
# example 2
print_first_last("Khulna University")
# example 3
print_first_last("Skill Morph")

# I completed this task (Task-9) with the help of ChatGPT.
```

```python
# Task-10 (Creating function for Uppercase string)
def to_uppercase(text):
    the_text = text.upper()
    print("Uppercase:", the_text)

# example 1
to_uppercase("Tasdid")
# example 2
to_uppercase("Khulna University")
# example 3
to_uppercase("Skill Morph")

# I completed this task (Task-10) with the help of ChatGPT.
```
