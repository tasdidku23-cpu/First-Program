## **Loop**

```python
# Looping through a list
fruits = ["apple", "banana", "orange"]
for fruit in fruits:
  print(f"I like {fruit}")
print("\n")
```

```python
# Practice
fruits = ("banana", "orange", "mango", "jackfruit")
for fruit in fruits:
  print(f"I hate {fruit}")
```

```python
# Using range() function
for i in range(5):  # 0 to 4
  print(i)
print("\n")
for i in range(2,10):  # 2 to 8
  print(i)
```

```python
# Practice
for i in range(11):
  print(i)
print("\n")
for i in range(5,9):
  print(i)
```

```python
# Continue until a condition is met
count = 0
while count < 5:
  print(count)
  count += 1  # Don't forget to increment!
print("\n")
# With a break statement
number = 0
while True:
  number += 1
  if number > 10:
    break
  print(number)
```

```python
# Continue statement - skip to next iteration
for num in range(10):
  if num % 2 == 0:
    continue  # Skip even numbers
  print(num)    # Prints only odd numbers
print("\n")
```

```python
# Break statement - exit the loop
for num in range(100):
  if num > 10:
    break
  print(num)
# Stop when we reach 11
# Prints 0 through 10
```

```python
# Multiplication table (1-5)
for i in range(1, 6):
  for j in range(1, 6):
    product = i * j
    print(f"{i} × {j} = {product}")
  print("-----")  # Separator between tables
```

```python
def calculate_stats(numbers):
  total = sum(numbers)
  average = total / len(numbers)
  maximum = max(numbers)
  minimum = min(numbers)
  return total, average, maximum, minimum
stats = calculate_stats([5, 10, 15, 20])
print(stats)  # (50, 12.5, 20, 5)
```

```python
total = 0  # Global variable

def add_to_total(value):
    global total  # Access global variable
    total += value
    return total

print(add_to_total(5))  # 5
print(add_to_total(10))  # 15
```
